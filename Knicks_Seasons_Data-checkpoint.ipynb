```python
pd.option_context('mode.use_inf_as_na', False)

# Define colors for each statistic
colors_knicks = {'3P%': 'green', 'TOV': 'red', 'AST': 'darkgreen', 'FG%': 'purple'}
colors_opponents = {'3P%': 'green', 'TOV': 'red', 'AST': 'darkgreen', 'FG%': 'purple'}

# Create a figure with grey background
fig, axes = plt.subplots(1, 2, figsize=(16, 8), facecolor='#F0F0F0')

# Plot Knicks' statistics
for col in ['3P%', 'TOV', 'AST', 'FG%']:
    sns.lineplot(data=All_Seasons_df, x='Season', y=col, marker='o', label=f'Knicks {col}', color=colors_knicks[col], ax=axes[0])

axes[0].set_title("Knicks' Per Season Stats")
axes[0].set_xlabel('Season')
axes[0].set_ylabel('Value')
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)

# Plot opponents' statistics
for col in ['3P%', 'TOV', 'AST', 'FG%']:
    sns.lineplot(data=Opponents_df, x='Season', y=col, marker='o', label=f'Opponent {col}', color=colors_opponents[col], ax=axes[1])

axes[1].set_title("Opponents' Per Season Stats")
axes[1].set_xlabel('Season')
axes[1].set_ylabel('Value')
axes[1].legend()
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
